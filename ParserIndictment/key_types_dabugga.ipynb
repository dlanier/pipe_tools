````python
import os
import sys
import numpy as np
import pandas as pd
import yaml
import json
from collections import defaultdict
````

````python
code_dir = '../../../Mayomics-Research/MayomicsVC/'
os.listdir(code_dir)
````

````python
#                               Open key_types.json
data_fullfilename = os.path.join(code_dir, 'key_types.json')

#                               key_types.json as text file
with open(data_fullfilename, 'r') as fh:
    lines = fh.readlines()
    
#                               key_types.json -- json read as python dict
with open(data_fullfilename, 'r') as fh:    
    json_dict = json.load(fh)
    
print(len(lines), 'lines found\n')

count = 0
#                               default dict of lines -- finds duplicates
lines_dict = defaultdict(int)
for line in lines:
    if '{' in line or '}' in line:
        pass
    else:
        if len(line.strip()) > 0:
            count += 1
            l = line.strip().split(":")
            lines_dict[l[0]] += 1

print(len(json_dict), 'dict lines')
print(len(lines_dict), 'lines_dict keys\n')

print('%i actual lines'%(count))
````

````python
#                               View the duplicates
for k, v in lines_dict.items():
    if v > 1:
        for line in lines:
            if not ('{' in line or '}' in line):
                if len(line.strip()) > 0:
                    line.replace('"', '')
                    line_list = line.strip().strip(',').split(":")
                    if len(line_list) > 1:
                        l_0 = line_list[0]
                        l_1 = line_list[1]
                        if l_0 == k:
                            print(k, l_1)
````

````python
#                               View the json dictionary of key_types.json
for k, v in json_dict.items():
    print('%30s: %s'%(k,v))
````

````python
#                               View the non-blank lines in key_types.json
for line in lines:
    if '{' in line or '}' in line:
        print(line.strip())
    else:
        if len(line.strip()) > 0:
            print('    ',line.strip())
````
